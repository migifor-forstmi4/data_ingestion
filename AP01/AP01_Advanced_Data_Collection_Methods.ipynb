#### <b>Aufgabe (2): Extrahieren Sie aus dem File 'BookStore.html' die Preise der Bücher mit Hilfe eines geeigneten XPath.</b>
<b>Details zur Aufgabenstellung:</b>
- Sie finden das HTML-File 'BookStore.html' im Order 'Data'.
- Komplettieren Sie für das Lösen der Aufgabe den unfertigen XPath im folgenden Python Code.
- Beachten Sie, dass nur ein einziger XPath für das Extrahieren aller Preise notwendig ist.
- Stellen Sie die extrahierten Preise im Jupyter Notebook dar.<br><br>
- Tipp: Für das Testen des XPath können Sie diesen Online XPath-Tester verwenden: http://xpather.com

<b style="color: gray;">(max. erreichbare Punkte: 4)</b>

```python
# Import File 'BookStore.html'
with open('BookStore.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parsen des HTML-Contents
tree = lxml_html.fromstring(html_content)

# XPath zum Extrahieren der Buchpreise (bitte dort wo im XPath '...' steht, komplettieren)
try:
    prices = tree.xpath('/html/body/div/section/div/div/p[2]/text()') #der xPath muss an die HTML Struktur der Datei angepasst werden; in diesem Beispiel sind es die Preise der Bücher.
    #mit section, div, p etc. werden die HTML Tags angesprochen - Die Angabe von Section ist relevevant um die Preise zu extrahieren, weil sie in diesem Fall in einem bestimmten Abschnitt der HTML Datei stehen
except:
    prices = [] # gibt eine leere Liste aus, falls der XPath nicht korrekt ist

# Anzeigen der extrahierten Buchpreise
print("Extracted Book Prices:", prices)
```

#### <b>Aufgabe (7): Beziehen Sie mit Hilfe der Overpass Turbo Web API Daten zu den Tankstellen in der Schweiz.</b>
<b>Details zur Aufgabenstellung:</b>
- Tankstellen sind mit dem key:value Paar "amenity"="fuel" in der OpenStreetMap Map-Feature Übersicht angegeben.
- Speichern Sie die Daten in einem File 'petrol_station.json' im Order 'Data' ab.
- Lesen Sie die Daten zusätzlich in ein pandas DataFrame und zeigen Sie dieses ganz oder in Teilen (z.B. erste 5 Zeilen) an.<br><br>
- Tipp: Auf der Webseite: https://overpass-turbo.eu können Sie Overpass Turbo Abfragen testen.

<b style="color: gray;">(max. erreichbare Punkte: 8)</b>

```python
# Overpass API URL
url = 'http://overpass-api.de/api/interpreter'

# Overpass turbo Abfrage-String
query = f"""
        [out:json];
        area["ISO3166-1"="CH"]["admin_level"="2"];
        node["amenity" = "fuel"] (area);
        out;"""


#out:json definiert das Ausgabeformat als JSON
#area["ISO3166-1"="CH"]["admin_level"="2"]; definiert die Abfrage auf das Gebiet der Schweiz (CH) auf admin_level 2 (Länderebene)
#node["amenity" = "fuel"] (area); sucht nach Knotenpunkten mit dem Tag "amenity" = "fuel" (Tankstellen) im definierten Gebiet
#out; gibt die Ergebnisse der Abfrage aus
        
# Web API Anfrage
r = requests.get(url, params={'data': query}) #--> sendet die GET-Anfrage an die Overpass API mit dem definierten Abfrage-String
data = r.json() ['elements'] #--> extrahiert die Elemente aus der JSON-Antwort der API

# Speichern der Daten in einer JSON-Datei
with open('petrol_station.json', 'w') as json_file: #--> definiert das Öffnen der Datei im Schreibmodus
    json.dump(data, json_file) #--> definiert das Speichern der Daten im JSON-Format in die Datei 'petrol_station.json'

# Speichern der Daten in einem Pandas DataFrame und Anzeigen des DataFrames
df = json_normalize(data) #--> json_normalize wandelt die verschachtelten JSON-Daten in ein flaches Pandas DataFrame um; ein flaches Pandas DataFrame ist eine tabellarische Darstellung der Daten, bei der alle verschachtelten Strukturen aufgelöst werden.
df.head(15) 

#Die Daten werden von der Overpass API abgerufen, die auf OpenStreetMap-Daten basiert.
```

#### <b>Aufgabe (8): Beziehen und Visualisieren Sie die aktuellen News-Feed (RSS-Feed) Daten der New York Times.</b>
<b>Details zur Aufgabenstellung:</b>
- Verwenden Sie diese URL zum RSS-Feed der New York Times Welt-Nachrichten: https://rss.nytimes.com/services/xml/rss/nyt/World.xml
- Beziehen Sie die RSS-Feeds und extrahieren sie die Titel der Artikel aus den RSS-Feed Einträgen.
- Zeigen Sie die Titel der RSS-Feed Einträge im Jupyter Notebook an.
- Erstellen Sie eine Graphik mit einer Word-Cloud der Titel der RSS-Feed Einträge und stellen Sie diese im Jupyter Notebook dar.

<b style="color: gray;">(max. erreichbare Punkte: 8)</b>

```python
# Parsen des RSS-News Feed von der New York Times World News Webseite
feed = feedparser.parse('https://rss.nytimes.com/services/xml/rss/nyt/World.xml') #--> feedparser.parse lädt und parst den RSS-Feed von der angegebenen URL

# Extrahieren der RSS-Feed Einträge
feed_entries = feed.entries #--> feed.entries enthält die Liste der Einträge im RSS-Feed; RSS bedeuetet Really Simple Syndication und ist ein Format zum Austausch von Nachrichten und anderen Inhalten im Web.

# Initialisieren des WordCloud-Texts
worldcloud_text = "" #--> definiert eine leere Zeichenkette für den WordCloud-Text

# Loop für Anzeige des Artikel-Titels
print("Article Titles:")
for entry in feed_entries:

    #Extrahieren des Artikel-Titels:
    article_title = entry.title #--> weist den Titel des aktuellen Eintrags der Variable article.title zu

    #Anhängen des Artikel-Titels an den WordCloud-Text:
    worldcloud_text += f"{article_title} " #--> fügt den Titel jedes Artikels zum WordCloud-Text hinzu owbei += bedeutet, dass der Text jedes Titels an die bestehende Zeichenkette worldcloud_text angehängt wird.

    #Anzeige des Artikel-Titels:
    print(f"{article_title}") #--> gibt den Titel des aktuellen Artikels aus / warum wird das f" vor dem String verwendet? --> f-Strings ermöglichen die Einbettung von Ausdrücken innerhalb von String-Literalen, indem sie ein f oder F vor den String setzen.

    #weil vorher F" mit geschweiften Klammern verwendet wurde, um Variablen in den String einzufügen, muss das bei der print Anweisung auch so gemacht werden.

# Generieren der WordCloud
worldcloud = WordCloud(width=700, height=400, background_color='black').generate(worldcloud_text) #--> erstellt eine WordCloud mit den angegebenen Abmessungen und Hintergrundfarbe aus dem worldcloud_text

# Anzeigen der WordCloud
plt.Figure(figsize=(10,6)) #--> definiert die Grösse der Abbildung
plt.imshow(worldcloud, interpolation='bilinear') #--> zeigt die WordCloud an, interpolation='bilinear' sorgt für eine glattere Darstellung der WordCloud
plt.axis('off') #--> entfernt die Achsenbeschriftungen
plt.show()
```
